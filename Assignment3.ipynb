```python
import numpy as py
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns

```

```python
ds = pd.read_excel("Heart.xlsx")
ds.head()
```

```python
ds.info()
```

```python
#Scatter Plot : Age vs Maxhr
plt.figure(figsize=(6, 6))
plt.scatter(ds['Age'], ds['MaxHR'], c='red', alpha=0.5)
plt.title('Scatter Plot: Age vs MaxHR')
plt.xlabel('Age')
plt.ylabel('MaxHR')
plt.grid(True)
plt.show()
```

```python
#Bar Plot Distribution of Heart Disease
plt.figure(figsize=(6, 6))
ds['AHD'].value_counts().plot(kind='bar', color=['blue', 'yellow'])
plt.title('Bar Plot: Distribution of Heart Disease')
plt.xlabel('Heart Disease (Yes/No)')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
#  Box Plot: Resting Blood Pressure (RestBP) distribution by Heart Disease
plt.figure(figsize=(6, 6))
ds.boxplot(column='RestBP', by='AHD', patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box Plot: RestBP Distribution by Heart Disease')
plt.suptitle('')
plt.xlabel('Heart Disease (Yes/No)')
plt.ylabel('Resting Blood Pressure')
plt.show()
```

```python
#  Pie Chart: Sex distribution in the dataset
sex_counts = ds['Sex'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(sex_counts, labels=['Male', 'Female'], autopct='%1.1f%%', startangle=90, colors=['lightblue', 'grey'])
plt.title('Pie Chart: Sex Distribution')
plt.show()
```

```python
# chest pain
ds.groupby('ChestPain').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Line Chart: Age vs Max Heart Rate (MaxHR)
plt.figure(figsize=(6, 6))
sns.lineplot(x='Age', y='MaxHR', hue='AHD', data=ds, marker='o', palette='coolwarm')
plt.title('Age vs Max Heart Rate (MaxHR) by Heart Disease (AHD)')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.show()
```
